```python
import pickle

with open(f"metrics_lfr_sf_50_generated_graphs.pickle","rb") as fopen:
    daux = pickle.load(fopen)
```

```python
import matplotlib.pyplot as plt

for graph in ['1-wiki-Vote', '2-soc-Epinions', '3-email-EuAll']:
    for metric in ['in_top']:
        xlimit = 100
        xs = [j for j in daux[graph]['SF'][metric].keys()]
        ys = [daux[graph]['SF'][metric][j] for j in xs]
        plt.plot(xs[:xlimit],ys[:xlimit],c='blue',label="SF, in_top")

        xs = [j for j in daux[graph]['LFR'][metric].keys()]
        ys = [daux[graph]['LFR'][metric][j] for j in xs]
        plt.title(graph)
        plt.plot(xs[:xlimit],ys[:xlimit],c='cyan',label="LFR, in_top")
        plt.ylim(0,105)
    
    for metric in ['equal']:
        xlimit = 100
        xs = [j for j in daux[graph]['SF'][metric].keys()]
        ys = [daux[graph]['SF'][metric][j] for j in xs]
        plt.plot(xs[:xlimit],ys[:xlimit],c='orange',label="SF, equal")

        xs = [j for j in daux[graph]['LFR'][metric].keys()]
        ys = [daux[graph]['LFR'][metric][j] for j in xs]
        plt.title(graph)
        plt.plot(xs[:xlimit],ys[:xlimit],c='darkred',label="LFR, equal")
        plt.ylim(0,105)
    plt.legend(loc=4)
    plt.xlabel("Percentage of nodes considered")
    plt.show()
    plt.clf()
```
