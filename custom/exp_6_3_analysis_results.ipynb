```python
import pickle
import matplotlib.pyplot as plt

path = 'train_no_train_data.pickle'


with open(path,'rb') as f:
    data = pickle.load(f)
```

```python
# Trained results 

ys = []
for net in data:
    v = [k['train']['kt'] for k in data[net]['r']['pred']]
    ys.append(v)
print(len(ys[0]))
plt.violinplot(ys)
plt.title("Trained models")
plt.xticks([1,2,3,4],['wiki-Vote', 'soc-Epinions' , 'email-EuAll','web-Google'])
plt.ylim([-1,1])
plt.ylabel("KT Score")
plt.show()
```

```python
# NO Trained results 

ys = []
for net in data:
    v = [k['no_train']['kt'] for k in data[net]['r']['pred']]
    ys.append(v)

plt.violinplot(ys)
plt.title("No trained models")
plt.xticks([1,2,3,4],['wiki-Vote', 'soc-Epinions' , 'email-EuAll','web-Google'])
plt.ylim([-1,1])
plt.ylabel("KT Score")
plt.show()
```
